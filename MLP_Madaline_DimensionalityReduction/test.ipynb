```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
train_size = 15000
```

```python
data_frame = pd.read_csv('House Sales.csv')
```

```python
data_frame = data_frame.drop("id",axis=1).drop('date', axis=1)
```

```python
len(data_frame.columns)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=19, activation='sigmoid', input_shape= (19,)))
model.add(tf.keras.layers.Dropout(0.4))
model.add(tf.keras.layers.Dense(units=10, activation='relu'))
model.add(tf.keras.layers.Dense(units=1))
model.compile(optimizer='Adam', loss='MSE', metrics=['mean_squared_logarithmic_error'])
model.summary()
```

```python
X_train = data_frame[:train_size]
Y_train = data_frame['price'][:train_size]
X_test = data_frame[train_size:]
Y_test = data_frame['price'][train_size:]
```

<h1>Train our data</h1>

```python
history = model.fit(X_train, Y_train, epochs=200)
```

```python
plt.plot(history.history['loss'][5:])
```

```python
model.predict(X_test)
```
