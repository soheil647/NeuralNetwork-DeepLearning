```python

import matplotlib.pyplot as plt
```

```python
def corrections(i, r1, r2, m):
    k = i-r1
    h = i-r2
    j = i+r1+1
    l = i+r2+1
    if k<0:
        k = 0
    if h<0:
        h = 0
    if l>m:
        l = m
    if j>m:
        j = m
    return k, h, j, l

def activation_fn(x):
    if x<0:
        return 0
    elif x>2:
        return 2
    else:
        return x
```

```python
x = [0.27, 0.35, 0.44, 0.58, 0.66, 0.77, 0.4, 0.32, 0.20, 0.15, 0.08]
r1 = 0
r2 = 12
c1 = 0.6
c2 = -0.4
t_max = 10
```

```python
legend_plt = []
print ('t = 0')
for i in range(len(x)):
    print ('x{} = {}'.format(i, x[i]))
plt.plot(x)
for t in range(t_max): #step 3 step 7 & 8
    x_old = x.copy() #step 6
    plt.plot(x_old)
    print ('t = {}'.format(t+1))
    legend_plt.append(t)
    for i in range(len(x_old)):
        k, h, j, l = corrections(i, r1, r2, len(x_old))
        sum1 = sum((x_old[k:j]))
        left = sum(x_old[h:k])
        right = sum((x_old[j:l]))
        sum2 = left + right
        x[i] = activation_fn((c1*sum1 + c2*sum2)) # step 4 & 5
        print ('x{} = {}'.format(i, x[i]))
plt.legend(legend_plt)
plt.xlabel("Index of X")
plt.ylabel("itiration number")
plt.show()
```
